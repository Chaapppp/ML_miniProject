```python
import pandas as pd
import numpy as np
import ast
```

```python
credits = pd.read_csv('dataset/tmdb_5000_credits.csv')
movies = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
movies = movies.merge(credits, left_on='title', right_on='title')
```

```python
movies.head(1)
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head(1)
```

```python
def convert(obj):
    try:
        return [i['name'] for i in ast.literal_eval(obj)]
    except:
        return []
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies['keywords']
```

```python
movies['cast'] = movies['cast'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)[:3]])  # Top 3 actors
movies['cast']
```

```python
movies['crew'] = movies['crew'].apply(lambda x: [i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director'])
movies['crew']
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x).lower())
```

```python
movies['tags']
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(titles, cosine_sim=cosine_sim):
    indices = []
    for title in titles:
        if title in movies['title'].values:
            indices.append(movies[movies['title'] == title].index[0])
    
    sim_scores = np.sum([cosine_sim[idx] for idx in indices], axis=0)  # Sum similarities
    sim_scores = list(enumerate(sim_scores))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Remove the input movies from recommendations
    sim_scores = [i for i in sim_scores if i[0] not in indices][:100]

    return movies['title'].iloc[[i[0] for i in sim_scores]].tolist()
```

```python
#Get user input
user_movies = []
print("Enter up to 5 movie titles (press Enter to stop):")
for i in range(5):
    title = input(f"Movie {i+1}: ").strip()
    if title:
        user_movies.append(title)
    else:
        break
```

```python
# Get recommendations
recommendations = get_recommendations(user_movies)
print("\nRecommended Movies:")
print(recommendations)
```

```python
import pickle
with open('movie_data.pkl', 'wb') as file:
    pickle.dump((movies, cosine_sim), file)
```
